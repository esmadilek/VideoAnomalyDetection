```python
# Sadece ilk veri ön işlemede bir kez çalıştırılır! - #3

#@title Import Train Dataset (Normal datasets only) - For Data Pre-Processing
def import_data (path):
    i = 0
    # loop over the training folders (Train01,Train02,..)
    for f in sorted(listdir(path)):
        directory_path = join(path, f)
        print("directory_path=", directory_path)
        if isdir(join(path, f)):
            print("# Importing " + directory_path)
            # loop over all the images in the folder (000001.jpg,..,001101.jpg)
            for c in sorted(listdir(join(path, f))):
                i = i + 1
                if str(join(join(path, f), c))[-3:] == "jpg":
                    img = Image.open(join(join(path, f), c))
                    img.save(paths.training_dataset_normal+"/{0:06}".format(i)+".jpg")
    print("Number of normal files: ", i)
    
print("Reading Images from folders :", paths.train)
import_data(paths.train)
print("Training Set is imported successfully \n")
```

```python
# Sadece ilk veri ön işlemede bir kez çalıştırılır! - #4

#@title Get Ground Truth Values

#### Get Ground Truth Values Start
def check_anomaly(dir_no, file_no):
  ground_truth = np.ones(20000, dtype = int)
  with open(paths.observed + '/observed.m') as f:
      lines = [line.rstrip('\n') for line in f]

  # r is a list(list)
  # each element of r is of form ['5:90', ' 140:200']
  r=[x.split(',') for x in lines]

  # val is a list(list(list))
  # each element of val is of form [['5', '90'], [' 140', '200']]
  val = list(map(lambda xs: [x.split(':') for x in xs], r))
  curr = val[dir_no-1]
  print("current=", curr)

  for x in curr:
    start = int(x[0])-1
    stop =  int(x[1])
    print("start=", start+1, " stop=",stop)
    for k in range(start, stop):
      ground_truth[k] = 0

  return ground_truth[file_no-1] #if normal, it returns 1; else if anomaly, it returns 0

#### Get Ground Truth Values Stop
```

```python
#@title Split Dataset into Training and Validation Datasets - #6
import shutil, random, os

srcPath = paths.training_dataset
destPath = paths.validation_dataset

print("srcPath", srcPath)
print("destPath", destPath)

def splitDataSet(dirPath, destDirectory, splitRatio):
  Labels = ['/Normal/','/Anomaly/']
  for label in Labels:
    samplePath = dirPath + label
    samples = os.listdir(samplePath)
    print('Processing folder --> ' + samplePath)
    print('Number of samples :' + str(len(samples)))
    savePath = destDirectory + label

    numberOfTestSamples = np.ceil(len(samples) * splitRatio).astype(int)
    print("numberOfTestSamples=", numberOfTestSamples)
    filenames = random.sample(samples, numberOfTestSamples)
    i = 0
    for fname in filenames:
      i += 1
      print("fname = ", fname)
      srcPath = os.path.join(samplePath, fname)
      destPath = os.path.join(savePath,fname)
      os.replace(srcPath, destPath)

    print("fileCount=", i)

splitDataSet(srcPath, destPath, paths.splitRatio)
```

```python
#@title Split Dataset into Validation and Test Datasets - #7

srcPath = paths.validation_dataset
destPath = paths.test_dataset

print("srcPath", srcPath)
print("destPath", destPath)

def splitDataSet(dirPath, destDirectory, splitRatio):
  Labels = ['/Normal/','/Anomaly/']
  for label in Labels:
    samplePath = dirPath + label
    Samples = os.listdir(samplePath)
    print('Processing folder --> ' + samplePath)
    print('Number of samples :' + str(len(Samples)))
    savePath = destDirectory + label

    numberOfTestSamples = np.ceil(len(Samples) * splitRatio).astype(int)
    filenames = random.sample(Samples, numberOfTestSamples)
    i = 0
    for fname in filenames:
      i += 1
      print("fname = ", fname, " i=",i)
      srcPath = os.path.join(samplePath, fname)
      destPath = os.path.join(savePath,fname)
      os.replace(srcPath, destPath)

    print("fileCount=", i)

splitDataSet(srcPath, destPath, paths.splitRatio)
```
